```python
df_fp = fpgrowth(df, min_support=0.3,use_colnames= True)
```

```python
df_gain20 = df.loc[(df["20dayGain"]==1)]
df_gain20.describe()
```

```python
df_nogain = df.loc[(df["5dayGain"] == 0)&(df["20dayGain"]==0)]
df_nogain.describe()
```

```python
df_gain5_des = df_gain5.describe()
df_gain20_des = df_gain20.describe()
df_nogain_des = df_nogain.describe()
for name in df.columns:
    if((df_gain5_des[name]["mean"] - df_nogain_des[name]["mean"])>0.1) or ((df_gain20_des[name]["mean"] - df_nogain_des[name]["mean"])>0.1):
        print(name)
```

```python
del df_gain5["5dayGain"],df_gain5["20dayGain"]
del df_gain20["20dayGain"],df_gain20["5dayGain"]
del df_nogain["5dayGain"],df_nogain["20dayGain"]
```

```python
patterns_5 = fpgrowth(df_gain5, min_support=0.35, use_colnames=True, max_len=None, verbose=0)
patterns_20 = fpgrowth(df_gain20, min_support=0.35, use_colnames=True, max_len=None, verbose=0)
patterns_ng = fpgrowth(df_nogain, min_support=0.2, use_colnames=True, max_len=None, verbose=0)
```

```python
def sig_patt(row,length,patterns_ng,df_nogain):
    pattern = row["itemsets"]
    confidence = -1.0
    for i,r in patterns_ng.iterrows():
        if (pattern == r["itemsets"]):#find the pattern if it's significant#
            confidence = (row["support"]*length/(row["support"]*length+r["support"]*len(df_nogain)))#Precision#
            #How many of those positive are negative that will hurt us badly#
            if confidence>=(length/(len(df_nogain)+length)):
                return confidence,True
        else:
            continue
    if confidence == -1.0:
        return 0.0,True
    else:
        return confidence,False
```

```python
day5=[]
length = len(df_gain5)
print(length/(len(df_nogain)+length))
for i,row in patterns_5.iterrows():
    conf,boo = sig_patt(row,length,patterns_ng,df_nogain)
    if boo:
        day5.append([conf,row["itemsets"]])
```

```python
day20=[]
length = len(df_gain20)
print(length/(len(df_nogain)+length))
for i,row in patterns_20.iterrows():
    conf,boo = sig_patt(row,length,patterns_ng,df_nogain)
    if boo:
        day20.append([conf,row["itemsets"]])
```
